## 2D Classical Ising Model

```python
# Read from file: 2DIsing_results.txt
# Plot T vs. Energy, Magnetization

import numpy as np
import matplotlib.pyplot as plt


L = 30
# Read from file
f = open("2DIsing_results.txt", "r")
lines = f.readlines()
f.close()

f = open("2DIsing_results_obc.txt", "r")
lines_obc = f.readlines()
f.close()

# Extract data
T = []
E = []
M = []

for line in lines:
    if line.split('=')[0] == 'T':
        T.append(float(line.split('=')[1]))
    elif line.split(' ')[0] == 'Energy':
        E.append(float(line.split(' ')[-1]))
    elif line.split(' ')[0] == 'Total':
        M.append(float(line.split(' ')[-1]))

T_obc = []
E_obc = []
M_obc = []

for line in lines_obc:
    if line.split('=')[0] == 'T':
        T_obc.append(float(line.split('=')[1]))
    elif line.split(' ')[0] == 'Energy':
        E_obc.append(float(line.split(' ')[-1]))
    elif line.split(' ')[0] == 'Total':
        M_obc.append(float(line.split(' ')[-1]))



# Plot
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].plot(T, E, 'o-')
# ax[0].plot(T_obc, E_obc, 'o-', label='OBC')
ax[0].set_xlabel('Temperature')
ax[0].set_ylabel('Energy')
ax[0].set_title('Energy vs. Temperature')

ax[1].plot(T, M, 'o-')
# ax[1].plot(T_obc, M_obc, 'o-', label='OBC')
ax[1].set_xlabel('Temperature')
ax[1].set_ylabel('Magnetization')
ax[1].set_title('Magnetization vs. Temperature')

ax[0].grid()
ax[1].grid()

# Plot the theoretical critical temperature
Tc = 2 / np.log(1 + np.sqrt(2))
ax[0].axvline(Tc, color='r', linestyle='--', label=r'Theoretical $T_c$')
ax[1].axvline(Tc, color='r', linestyle='--', label=r'Theoretical $T_c$')

# Add title
fig.suptitle(rf'MC Simulation of {L}x{L} 2D Classical Ferromagnetic Ising Model ($k_b=1, J=1$)')

plt.tight_layout()
ax[0].legend()
ax[1].legend()
plt.show()
```
